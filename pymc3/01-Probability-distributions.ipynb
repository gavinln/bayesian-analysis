# Probability distributions

```python
import pymc3 as pm
import seaborn as sns
```

```python
with pm.Model():
    n = pm.distributions.Normal('n', mu=0, sigma=1)
    b = pm.distributions.Beta('b', alpha=1, beta=4)
    br = pm.distributions.Bernoulli('br', p=0.4)
    bi = pm.distributions.Binomial('bi', n=4, p=0.75)
    p = pm.distributions.Poisson('p', mu=0.5)
    e = pm.distributions.Exponential('e', lam=0.5)
```

```python
with pm.Model():
    n_samples = n.random(size=10000)
    b_samples = b.random(size=10000)
    br_samples = br.random(size=10000)
    bi_samples = bi.random(size=10000)
    p_samples = p.random(size=10000)
    e_samples = p.random(size=10000)
```

```python
g = sns.displot(n_samples)
ax = g.axes[0][0]
ax.set_title('Normal distribution')
```

```python
g = sns.displot(b_samples, kind='hist')
ax = g.axes[0][0]
ax.set_title('Beta distribution')
```

```python
g = sns.displot(br_samples, kind='hist')
ax = g.axes[0][0]
ax.set_title('Bernoulli distribution')
```

```python
g = sns.displot(bi_samples, kind='hist')
ax = g.axes[0][0]
ax.set_title('Binomial distribution')
```

```python
g = sns.displot(p_samples, kind='hist')
ax = g.axes[0][0]
ax.set_title('Poisson distribution')
```

```python
g = sns.displot(e_samples, kind='hist')
ax = g.axes[0][0]
ax.set_title('Exponential distribution')
```
